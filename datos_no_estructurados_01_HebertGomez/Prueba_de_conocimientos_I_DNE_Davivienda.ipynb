```python
import os
import cv2
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from skimage.feature import hog
import pandas as pd
# Definir rutas
input_folder = "/content/sample_data/images/"                                  #todas
output_folder_with_content = "/content/output_with_content/"
output_folder_without_content = "/content/output_without_content/"

# Crear carpetas de salida si no existen
os.makedirs(output_folder_with_content, exist_ok=True)
os.makedirs(output_folder_without_content, exist_ok=True)

# Preprocesamiento de imágenes y extracción de características
def preprocess_and_extract_features(img_path, img_size=(128, 128)):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        return None
    img = cv2.resize(img, img_size)
    img = img / 255.0
    hog_features = hog(img, pixels_per_cell=(8, 8), cells_per_block=(2, 2), feature_vector=True)

    return hog_features

# Cargar y preprocesar las imágenes
image_paths = [os.path.join(input_folder, filename) for filename in os.listdir(input_folder) if filename.endswith(('.png'))]
features = [preprocess_and_extract_features(img_path) for img_path in image_paths]
features = [f for f in features if f is not None]
#Reducción de dimensionalidad para acelerar el clustering
pca = PCA(n_components=50)
reduced = pca.fit_transform(features)

# Aplicar K-means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(features)
labels = kmeans.labels_
print(labels)

```

```python
for img_path, label in zip(image_paths, labels):
    if label == 1:
        destination = output_folder_with_content
    else:
        destination = output_folder_without_content
    os.rename(img_path, os.path.join(destination, os.path.basename(img_path)))
```
